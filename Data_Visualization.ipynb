```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(f'Pandas:', pd.__version__)
print(f'Numpy:', np.__version__)
print(f'Matplotlib:', matplotlib.__version__)
```

```python
plt.style.use('seaborn')
```

```python
df_iris = pd.read_csv("iris.csv", names=['sepal_length',
                                          'sepal_width', 
                                          'petal_length',
                                          'petal_width',
                                          'species'])
df_iris.head()
```

```python
print('Total rows:', df_iris.shape[0])
print('Total column:', df_iris.shape[1])
```

### Bar Chart

```python
#comparison
df_iris['species'].hist()
plt.show()
```

### Scatter Plot

```python
df_iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
df_iris.plot(kind='scatter', x='petal_length', y='petal_width')
plt.show()
```

### Distplot

```python
sns.distplot(df_iris['sepal_length'], bins=20)
plt.show()

#bins digunakan untuk memperkecil jarak data agar lebih detail
```

### Box Plot

```python
sns.boxplot(data=df_iris)
plt.xlabel('Features')
plt.ylabel('Length in cm')
plt.show()
#garis bawah menunjukkan data min, garis atas menunjukkan data max
#titik maksudnya data outlier
```

```python
sns.boxplot(data=df_iris, x='species', y='petal_length')
plt.xlabel('Species')
plt.ylabel('Length in cm')
plt.show()
```

### Pairplot

```python
sns.pairplot(data=df_iris, hue='species')
plt.show()

#hue berguna untuk memberikan warna sesuai kolom yg dipilih
```

### Heatmap

```python
sns.heatmap(data=df_iris.corr(), annot=True)
plt.show()

#untuk mengetahui korelasi antar data
```
