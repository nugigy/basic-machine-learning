<b> Tugas 1: Data Visualization menggunakan datasets titanic.csv untuk menganalisa informasi <i>Single Variable, Two Variable</i> dan <i>Multi Variable</i>!</b>

#### Library

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Dataset

```python
data = pd.read_csv("titanic.csv")
data.head()
```

```python
data.info()
```

### Single Variable

```python
#mengetahui jumlah penumpang yang selamat(1) dan meninggal(0)

data['Survived'].hist()
plt.show()
```

```python
#mengetahui persentase jumlah penumpang berdasarkan pemilihan class

plt.pie([data['Pclass'][data['Pclass']==1].count(), data['Pclass'][data['Pclass']==2].count(),
         data['Pclass'][data['Pclass']==3].count()], 
        labels=['Kelas 1', 'Kelas 2', 'Kelas 3'], 
        colors=['salmon', 'khaki', 'yellowgreen'],
        autopct='%1.1f%%')
```

```python
#mengetahui sebaran jumlah penumpang berdasarkan usia

sns.boxplot(x=data["Age"])
```

```python
#mengetahui sebaran jumlah penumpang berdasarkan usia

sns.distplot(data['Age'], bins=25)
plt.show()
```

### Double Variable

```python
#mengetahui jumlah penumpang selamat/tidak berdasarkan pemilihan class

sns.countplot('Survived', hue='Pclass', data=data)
```

```python
#mengetahui jumlah penumpang laki-laki&perempuan berdasarkan selamat/tidak

sns.countplot('Sex', hue='Survived', data=data)
```

```python
#mengetahui sebaran data antara usia dengan biaya

data.plot(kind='scatter', x='Age', y='Fare')
plt.show()
```

### Multi Variable

```python
#membandingkan jumlah penumpang perempuan&laki-laki yang selamat/tidak berdasarkan pemilihan kelas

sns.barplot(x='Sex', y='Survived', hue='Pclass', data=data)
```

```python
#menggambarkan korelasi antar kolom

sns.pairplot(data=data, hue='Survived')
plt.show()
```

```python
#mengetahui nilai korelasi antar kolom

sns.heatmap(data=data.corr(), annot=True)
plt.show()
```
