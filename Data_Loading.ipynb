#### Change Columns

```python
#merubah nama kolom dari data
new_columns = {'Survived': 'SURVIVED', 
 'Pclass': 'PCLASS', 
 'Name': 'NAME', 
 'Sex': 'SEX', 
 'Age': 'AGE', 
 'Siblings/Spouses Aboard': 'SIBSA', 
 'Parents/Children Aboard': 'PARCA' , 
 'Fare': 'FARE'}

df_titanic = df_titanic.rename(columns=new_columns)
df_titanic
```

#### Check Value

```python
#mencari tahu data kosong dalam dataframe
df_titanic.isnull()
```

```python
#mencari tahu data kosong berdasarkan kolom, (false=tidak ada data kosong, true=ada data kosong)
df_titanic.isnull().any()
```

```python
#mencari data berdasarkan index (urutan baris keberapa)
count = 0
for x, row in df_titanic.iterrows():
    print('Index:', x)
    print(row)
    count += 1
    
    if count == 3:
        break
```

#### Group Value

```python
#mengetahui rata-rata berdasarkan pclass di data
df_titanic.groupby('PCLASS').mean()
```

```python
#mengetahui jumlah/sum dari data berdasarkan pclass
df_titanic.groupby('PCLASS').sum().reset_index().head()
```

```python
#menggambarkan jumlah/sum dari data berdasarkan pclass ke dalam grafik
df_titanic.groupby('PCLASS').sum().reset_index().head().plot(kind='bar')
```

#### Concatenate Data

```python
#memisahkan beberapa kolom menjadi dataframe baru
df_titanic1 = df_titanic[['SURVIVED', 'PCLASS', 'NAME']]
df_titanic2 = df_titanic[['SEX', 'AGE', 'SIBSA']]
df_titanic3 = df_titanic[['PARCA', 'FARE']]
```

```python
#menggabungkan beberapa dataframe menjadi dataframe baru
frames = [df_titanic1, df_titanic2, df_titanic3]
df_titanic_concatenated = pd.concat(frames, axis=1)
df_titanic_concatenated.head()
```
